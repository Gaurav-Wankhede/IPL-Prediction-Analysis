```python
import glob
import pandas as pd

# Define the path to the csv folder
csv_folder_path = 'csv'

# Use glob to find all the csv files in the folder
csv_files = glob.glob(f'{csv_folder_path}/*.csv')

# Check if any csv files were found
if not csv_files:
    print('Error: No CSV files found in the specified folder.')
else:
    # Initialize an empty list to store the dataframes
    dfs = []

    # Loop through the csv files and append each dataframe to the list
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)

        # Rename the columns in each dataframe to make them consistent
        if 'Batting_ID' in df.columns:
            df = df.rename(columns={'Batting_ID': 'ID', 'Innings': 'Inning', 'Venue': 'Venue', 'Team': 'Team', 'Start Date': 'Start_Date', 'End Date': 'End_Date', 'BATTING': 'Player', 'Dismissal type': 'Dismissal_Type', 'R': 'Runs', 'B': 'Balls', 'M': 'Minutes', '4s': 'Fours', '6s': 'Sixes', 'SR': 'Strike_Rate'})
        elif 'Bowling_Player' in df.columns:
            df = df.rename(columns={'Inning': 'Inning', 'Venue': 'Venue', 'Team': 'Team', 'Start Date': 'Start_Date', 'End Date': 'End_Date', 'Bowling_Player': 'Player', 'Wickets': 'Wickets', 'M': 'Maidens', 'R': 'Runs', 'W': 'Wides', 'NB': 'No_Balls', 'ECON': 'Economy', '0s': 'Dot_Balls', '4s': 'Fours', '6s': 'Sixes'})
        elif 'Innings' in df.columns and 'Team1' in df.columns:
            df = df.rename(columns={'Innings': 'Inning', 'Venue': 'Venue', 'Start_Date': 'Start_Date', 'End_Date': 'End_Date', 'Team1': 'Team', 'Team1_runs': 'Runs', 'Team1_wickets': 'Wickets', 'Team2_runs': 'Target', 'Team1_played_Overs': 'Played_Overs', 'Team1_Max_Overs': 'Max_Overs', 'Match_Result': 'Result'})
        elif 'MVP_ID' in df.columns:
            df = df.rename(columns={'MVP_ID': 'ID', 'Innings': 'Inning', 'Venue': 'Venue', 'Start_Date': 'Start_Date', 'End_Date': 'End_Date', 'Player_Name': 'Player', 'Team': 'Team', 'Total_Impact': 'Total_Impact_Points', 'Runs': 'Runs', 'Impact_Runs': 'Impact_Runs_Points', 'Batting_Impact': 'Batting_Impact_Points', 'Bowl': 'Wickets', 'Impact_Wickets': 'Impact_Wicket_Points', 'Bowling_Impact': 'Bowling_Impact_Points', 'Balls_Faced': 'Balls_Faced', 'Runs_Conceded': 'Runs_Conceded'})
        elif 'Over_ID' in df.columns:
            df = df.rename(columns={'Over_ID': 'ID', 'Innings': 'Inning', 'Venue': 'Venue', 'Start_Date': 'Start_Date', 'End_Date': 'End_Date', 'Team_Name': 'Team', 'Overs': 'Overs', 'Team_Runs': 'Runs', 'Team_Wickets': 'Wickets', 'Team_Total_Runs': 'Total_Runs', 'Team_Total_Wickets': 'Total_Wickets'})
        elif 'Innings' in df.columns and 'Player1' in df.columns:
            df = df.rename(columns={'Innings': 'Inning', 'Venue': 'Venue', 'Start_Date': 'Start_Date', 'End_Date': 'End_Date', 'Team': 'Team', 'Player1': 'Player1', 'Player1_Runs': 'Runs1', 'Player1_Balls': 'Balls1', 'Partnership_Runs': 'Partnership_Runs', 'Partnership_Balls': 'Partnership_Balls', 'Player2_Runs': 'Runs2', 'Player2_Balls': 'Balls2'})

        dfs.append(df)

    # Merge the first two dataframes in the list
    merged_df = pd.merge(dfs[0], dfs[1], on=['Inning', 'Venue', 'Start_Date', 'End_Date'])
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('./database/IPL_Prediction_Analysis.db')

# Query the Batting, Bowling, Match_Data, MVP_Data, Overs_Data, and Stats_Data tables
batting_query = '''
SELECT Innings, Venue, Start_Date, End_Date, Player_name, Dismissal_type, Runs, Balls, Minutes, Fours, Sixes, Strike_Rate
FROM Batting

'''
bowling_query = '''
SELECT Innings, Venue, Start_Date, End_Date, Bowling_Player, Overs, Maidens, Runs, Wickets, Economy_rate, Dot_ball, Fours, Sixes, Wides, No_balls
FROM Bowling

'''
match_query = '''
SELECT Innings, Venue, Start_Date, End_Date, Team1, Team2, Team1_runs, Team1_wickets, Team2_runs, Team2_wickets, Target, Team1_played_Overs, Team2_played_Overs, Team1_Max_Overs, Team2_Max_Overs, Match_Result
FROM Match_Data

'''
mvp_query = '''
SELECT Innings, Venue, Start_Date, End_Date, Player_Name, Team, Total_Impact, Runs, Balls_Faced, Impact_Runs, Batting_Impact, Wickets, Runs_Conceded, Impact_Wickets, Bowling_Impact
FROM MVP_Data

'''
overs_query = '''
SELECT Innings, Venue, Start_Date, End_Date, Team_Name, Overs, Team_Runs, Team_Wickets, Team_Total_Runs, Team_Total_Wickets
FROM Overs_Data

'''
stats_query = '''
SELECT Innings, Venue, Start_Date, End_Date, Team, Player1, Player2, Player1_Runs, Player1_Balls, Partnership_Runs, Partnership_Balls, Player2_Runs, Player2_Balls
FROM Stats_Data

'''

# Read the queries into pandas dataframes
batting_df = pd.read_sql_query(batting_query, conn)
bowling_df = pd.read_sql_query(bowling_query, conn)
match_df = pd.read_sql_query(match_query, conn)
mvp_df = pd.read_sql_query(mvp_query, conn)
overs_df = pd.read_sql_query(overs_query, conn)
stats_df = pd.read_sql_query(stats_query, conn)

# Merge the dataframes on the common columns
merged_df = pd.concat([batting_df, bowling_df, match_df, mvp_df, overs_df, stats_df], axis=1)

# Reset the index and create a new ID column
merged_df.reset_index(drop=True, inplace=True)
merged_df['Match_ID'] = merged_df.index + 1

# Save the merged data as a CSV file
merged_df.to_csv('merged_data.csv', index=False)

print(merged_df,"\n\nmerged_data.csv is created successfully")
# Close the database connection
conn.close()
```
